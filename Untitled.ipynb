```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import IsolationForest
```

```python
#User 1
user_train = pd.read_csv('User1/train/user_1.csv')
user_test = pd.read_csv('User1/test/user_1.csv')

y_train = user_train['class']
y_test = user_test['class']
user_train = user_train.drop(['class'], axis=1)
user_test = user_test.drop(['class'], axis=1)
X_train = user_train.to_numpy()
X_test = user_test.to_numpy()

print("Size of train data = ", np.shape(X_train))
print("Size of test data = ", np.shape(X_test))

#SVM
#clf = svm.SVC(kernel='linear', C=10, random_state=14)
clf = svm.LinearSVC()
print("Training begins ............")
clf.fit(X_train, y_train)
print("Training completed")
print("Prediction started")
y_pred = clf.predict(X_test)
print("Evaluation")
print("Confusion matrix is \n:", confusion_matrix(y_test, y_pred))
print("Accuracy score = {:.2f}".format(accuracy_score(y_test, y_pred)*100))
print("Precision = {:.4f}".format(precision_score(y_test, y_pred)))
print("Recall = {:.4f}".format(recall_score(y_test, y_pred)))
print("F1-score = {:.4f}".format(f1_score(y_test, y_pred)))
```

```python
#User 1
user_train = pd.read_csv('User3/train/user_3.csv')
user_test = pd.read_csv('User3/test/user_3.csv')

y_train = user_train['class']
y_test = user_test['class']
user_train = user_train.drop(['class'], axis=1)
user_test = user_test.drop(['class'], axis=1)
X_train = user_train.to_numpy()
X_test = user_test.to_numpy()


m,n  = np.shape(X_train)
rand_mat = np.random.randn(n, 4*n)
Xtr = np.matmul(X_train, rand_mat)
Xte = np.matmul(X_test, rand_mat)

X_tr = np.concatenate((np.cos(Xtr), np.sin(Xtr)), axis=1)
X_te = np.concatenate((np.cos(Xte), np.sin(Xte)), axis=1)

print("Size of train data = ", np.shape(X_tr))
print("Size of test data = ", np.shape(X_te))

#SVM
clf = svm.SVC(kernel='linear', C=10, random_state=14)
print("Training begins ............")
clf.fit(X_tr, y_train)
print("Training completed")
print("Prediction started")
y_pred = clf.predict(X_te)
print("Evaluation")
print("Confusion matrix is \n:", confusion_matrix(y_test, y_pred))
print("Accuracy score = {:.2f}".format(accuracy_score(y_test, y_pred)*100))
print("Precision = {:.4f}".format(precision_score(y_test, y_pred)))
print("Recall = {:.4f}".format(recall_score(y_test, y_pred)))
print("F1-score = {:.4f}".format(f1_score(y_test, y_pred)))
```
